## Q10:

```python
import time
from playsound import playsound

def countdown(time_sec):
    while time_sec:
        mins, secs = divmod(time_sec, 60)
        hrs = int(mins/60)
        timeformat = '{:02d}:{:02d}:{:02d}'.format(hrs, mins ,secs)
        print(timeformat, end='\r')
        time.sleep(1)
        time_sec -= 1

    playsound("D:\\extra-ded\\projects\\python\\timer\\nuke.mp3")
    print("stop")

secs = int(input("enter the secs you want:"))
countdown(secs)
```
